# CNN version Laser Machine Listener: Training

Two training approaches were made here; training from scratch and transfer learning.

This problem has very small dataset, then usually there's no option to train from scratch for CNN model.
Training from scratch is just experimental attempt, transfer learning should be employed for similar problems.

```python
import sys
sys.path.append('../..')
from lib_train import *
%matplotlib inline

conf.folder = Path('.')
conf.test_size = 0.2

print(conf)

# Load all dataset -> X|y|idx_train, X|y|idx_valid
X_train, y_train, idx_train = load_dataset('X_train.npy', 'y_train.npy', 'idx_train.npy')
X_valid, y_valid, idx_valid = load_dataset('X_valid.npy', 'y_valid.npy', 'idx_valid.npy')
print('Loaded train:valid = %d:%d samples.' % (len(X_train), len(X_valid)))

# Normalize samplewise if requested
if conf.normalize == 'samplewise':
    print(' normalize samplewise')
    samplewise_mean_X(X_train)
    samplewise_mean_X(X_valid)
```

## Train from scratch

Basically this approach doesn't work.
CNN is supposed to acquire good feature representation capability in convolutional layers from large amount of data, which cannot be extracted from few data.

Though this training finally yields a good model, but this is just by luck and techniques. As you can see with `val_acc`, training didn't converged throughout epochs due to too few source of data.

```python
conf.epochs = 500
conf.best_weight_file = 'model_from_scratch.h5'
history, model, plain_datagen = train_model(conf, fold=0,
                                            dataset=[X_train, y_train, idx_train], 
                                            model=None, init_weights=None)
acc = evaluate_model(conf, model, plain_datagen, X_valid, y_valid) 
print('___ training finished ___')
```

## Train with pre-trained model, transfer learning

Freesound Dataset Kaggle 2018 pretrained model is transferred to this problem. It converges fast, and the accuracy is fine thanks to the pretrained better representation.

```python
import keras.backend as K
K.clear_session()

conf.best_weight_file = 'model_based_on_fsd_model.h5'
conf.epochs = 100
history, model, plain_datagen = train_model(conf, fold=0,
                                            dataset=[X_train, y_train, idx_train], 
                                            model=None, 
                                            init_weights='../../model/mobilenetv2_small_fsd2018_41cls.h5')
acc = evaluate_model(conf, model, plain_datagen, X_valid, y_valid) 
print('___ training finished ___')
```
